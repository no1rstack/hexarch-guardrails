## 🎯 Demo 1: Rate Limiting Protection

Let's create a function that would normally spam an API, but Hexarch will block it!

````python
import yaml
from hexarch_guardrails import Guardian
import time
from pathlib import Path

# Create a policy file
policy_config = {
    'policies': [
        {
            'id': 'rate_limit',
            'description': 'Prevent API spam',
            'rules': [
                {
                    'resource': 'demo_api',
                    'requests_per_minute': 5,
                    'action': 'block'
                }
            ]
        }
    ]
}

# Write policy file
Path('hexarch.yaml').write_text(yaml.dump(policy_config))

# Initialize guardian
guardian = Guardian()

# Simulated API call
@guardian.check('rate_limit')
def call_expensive_api(message):
    """Simulates calling an expensive API"""
    print(f"  ✅ API call successful: {message}")
    return {"status": "success", "message": message}

print("\n🔥 Attempting to spam the API with 10 rapid calls...\n")

success_count = 0
blocked_count = 0

for i in range(10):
    try:
        call_expensive_api(f"Request #{i+1}")
        success_count += 1
    except Exception as e:
        print(f"  🛑 Request #{i+1} BLOCKED: {str(e)[:50]}...")
        blocked_count += 1
    time.sleep(0.1)

print(f"\n📊 Results:")
print(f"   ✅ Allowed: {success_count}/10")
print(f"   🛑 Blocked: {blocked_count}/10")
print(f"\n💡 Hexarch protected you from {blocked_count} unnecessary API calls!")
````

## 💰 Demo 2: Budget Protection

Prevent runaway costs from AI API calls!

````python
# Update policy for budget protection
policy_config = {
    'policies': [
        {
            'id': 'ai_budget',
            'description': 'Prevent AI API overspending',
            'rules': [
                {
                    'resource': 'openai',
                    'monthly_budget': 50,
                    'action': 'warn_at_80%'
                }
            ]
        }
    ]
}

Path('hexarch.yaml').write_text(yaml.dump(policy_config))
guardian = Guardian()

# Track simulated spending
total_cost = 0.0

@guardian.check('ai_budget')
def call_gpt4(prompt, estimated_cost=2.50):
    """Simulates calling GPT-4 with cost tracking"""
    global total_cost
    total_cost += estimated_cost
    print(f"  💬 GPT-4 Response: <Generated text for: {prompt[:30]}...>")
    print(f"  💵 Cost: ${estimated_cost:.2f} | Total: ${total_cost:.2f}")
    return {"response": "Generated text", "cost": estimated_cost}

print("\n💸 Simulating AI API usage...\n")

prompts = [
    "Write a blog post about AI",
    "Generate product descriptions",
    "Create marketing copy",
    "Write documentation"
]

for idx, prompt in enumerate(prompts, 1):
    print(f"\n🔄 Request {idx}/4:")
    try:
        call_gpt4(prompt, estimated_cost=12.50)
    except Exception as e:
        print(f"  ⚠️  WARNING: {str(e)[:80]}")

print(f"\n\n💡 Budget alerts help you stay within spending limits!")
print(f"   Total spent this demo: ${total_cost:.2f}")
````

## 🔒 Demo 3: Dangerous Operation Protection

Require confirmation before destructive operations!

````python
# Policy for destructive operations
policy_config = {
    'policies': [
        {
            'id': 'safe_delete',
            'description': 'Protect against accidental deletions',
            'rules': [
                {
                    'operation': 'delete',
                    'action': 'require_confirmation'
                }
            ]
        }
    ]
}

Path('hexarch.yaml').write_text(yaml.dump(policy_config))
guardian = Guardian()

@guardian.check('safe_delete')
def delete_user_data(user_id, confirmed=False):
    """Dangerous operation that needs confirmation"""
    if not confirmed:
        raise Exception("⚠️  Deletion requires explicit confirmation!")
    print(f"  🗑️  User {user_id} data deleted (simulation)")
    return {"status": "deleted", "user_id": user_id}

print("\n🔥 Attempting to delete user data without confirmation...\n")

try:
    delete_user_data("user_12345")
except Exception as e:
    print(f"  🛑 BLOCKED: {e}")

print("\n✅ Now with explicit confirmation...\n")
try:
    delete_user_data("user_12345", confirmed=True)
    print("  ✅ Allowed: Explicit confirmation provided")
except Exception as e:
    print(f"  ❌ Error: {e}")

print("\n💡 Guardrails prevent accidental destructive operations!")
````
